## ¿Qué es `print()`?

`print()` es una función de Python que muestra información en pantalla.
Es como decirle al computador: "muéstrame esto".
La usaremos constantemente para ver resultados mientras aprendemos.


## Ejemplo guiado


````python
print("Bienvenido al curso de Introducción a la Programación")
resultado = 7 + 5
print("7 + 5 =", resultado)
````

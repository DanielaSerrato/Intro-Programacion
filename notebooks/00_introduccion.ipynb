## Tu primer `print()`

`print()` es una función que muestra información en pantalla.
Es la forma más simple de ver resultados mientras aprendes.


````python
print("¡Hola, mundo!")
print("Este es mi primer programa en Python")
````

## Variables y tipos básicos

Una **variable** guarda un valor con un nombre.
Los tipos básicos que veremos ahora son: **texto** (`str`), **números** (`int` y `float`) y **booleanos** (`bool`).


````python
nombre = "Ana"
edad = 20
altura = 1.65
activo = True

print(nombre)
print(edad)
print(altura)
print(activo)
````

## Primera función en Python

Una función agrupa instrucciones con un nombre.
Puede recibir parámetros y devolver un resultado con `return`.


````python
def saludo(nombre):
    return f"Hola, {nombre}!"

print(saludo("Carlos"))
````

### Ejercicio 1: Primer mensaje
**Enunciado:** crea una variable `mensaje` con un saludo y muéstrala en pantalla.

**Ejemplo esperado:**
```
Hola, soy estudiante de Python
```

**Pista:** usa `print(mensaje)`.


````python
# TU CÓDIGO AQUÍ
mensaje = ""

print(mensaje)

# Prueba simple
assert isinstance(mensaje, str) and mensaje != ""
````

### Ejercicio 2: Variables con tipo
**Enunciado:** crea tres variables: `nombre` (texto), `edad` (entero) y `ciudad` (texto).
Muestra una frase completa con esas variables usando un f-string.

**Ejemplo esperado:**
```
Sofía tiene 22 años y vive en Lima.
```

**Pista:** usa `f"{nombre} tiene {edad} años y vive en {ciudad}."`.


````python
# TU CÓDIGO AQUÍ
nombre = ""
edad = 0
ciudad = ""

frase = f"{nombre} tiene {edad} años y vive en {ciudad}."
print(frase)

# Prueba simple
assert isinstance(frase, str) and nombre in frase
````

### Ejercicio 3: Booleanos
**Enunciado:** crea una variable booleana `le_gusta_python` y muéstrala.

**Ejemplo esperado:**
```
True
```

**Pista:** los booleanos pueden ser `True` o `False`.


````python
# TU CÓDIGO AQUÍ
le_gusta_python = True
print(le_gusta_python)

# Prueba simple
assert isinstance(le_gusta_python, bool)
````

### Ejercicio 4: Tu primera función
**Enunciado:** crea una función `doble` que reciba un número y devuelva el doble.

**Ejemplo esperado:**
```
4 -> 8
```

**Pista:** recuerda usar `return`.


````python
# TU CÓDIGO AQUÍ

def doble(numero):
    return numero * 2

# Pruebas simples
assert doble(4) == 8
assert doble(0) == 0
````

# Algoritmos

En este capítulo aprenderás qué es un **algoritmo** y cómo describirlo con pasos claros.
Antes de programar, es útil ordenar una idea en instrucciones simples.
Eso evita errores y hace que el código sea más fácil de escribir.


## ¿Qué es un algoritmo?

Un algoritmo es una **secuencia ordenada de pasos** para resolver un problema.
No necesita código: puede ser una lista de instrucciones escrita con palabras simples.


## Ejemplo guiado

**Problema:** saludar a una persona con su nombre.

**Pasos del algoritmo:**
1. Pedir el nombre.
2. Guardar el nombre.
3. Mostrar un saludo con el nombre.


````python
# Una forma simple de representar pasos usando variables
paso1 = "Pedir el nombre"
paso2 = "Guardar el nombre"
paso3 = "Mostrar el saludo"

print(paso1)
print(paso2)
print(paso3)
````

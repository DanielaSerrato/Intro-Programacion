# Algoritmos


Este capítulo introduce el concepto de algoritmo como una secuencia ordenada de pasos.
Aprenderás a describir procesos de forma clara antes de programarlos.
Verás cómo pasar de un problema cotidiano a instrucciones precisas.
Es una base esencial para los siguientes capítulos del curso.


## Ejemplo guiado

**Pseudocódigo:**
INICIO
PASO 1: Pedir el nombre
PASO 2: Crear el saludo
PASO 3: Mostrar el mensaje
FIN


````python
pasos = ["Pedir el nombre", "Crear el saludo", "Mostrar el mensaje"]
for paso in pasos:
    print(paso)
````

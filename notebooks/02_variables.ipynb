## ¿Qué es una variable?

Una variable es un nombre que guarda un valor para poder usarlo después.
Es como una caja con etiqueta: puedes guardar un dato y consultarlo cuando lo necesites.
En Python el tipo puede cambiar porque el lenguaje es de tipado dinámico.


## Ejemplo guiado


````python
nombre = "Daniela"
edad = 29
activo = True

print(nombre, edad, activo)
````

### Ejercicio 1: Crear variables
**Enunciado:** crea variables `nombre`, `edad` y `ciudad` con valores apropiados.
Luego imprime una frase que combine las tres variables.

**Ejemplo esperado:**
```
María tiene 24 años y vive en Quito.
```

**Pista:** usa un f-string.


````python
# TU CÓDIGO AQUÍ
nombre = ""
edad = 0
ciudad = ""

frase = f"{nombre} tiene {edad} años y vive en {ciudad}."
print(frase)

# Prueba simple
assert isinstance(frase, str) and nombre in frase
````

### Ejercicio 2: Actualizar un valor
**Enunciado:** crea una variable `edad` y súmale 1. Luego imprime el nuevo valor.

**Ejemplo esperado:**
```
21
```

**Pista:** puedes usar `edad += 1`.


````python
# TU CÓDIGO AQUÍ
edad = 0
edad += 1
print(edad)

# Prueba simple
assert edad >= 1
````

### Ejercicio 3: Tipos básicos
**Enunciado:** crea una variable de cada tipo básico: `texto` (str), `entero` (int), `decimal` (float) y `activo` (bool).
Imprime cada una en una línea distinta.

**Ejemplo esperado:**
```
Hola
10
3.5
False
```

**Pista:** usa `print()` cuatro veces.


````python
# TU CÓDIGO AQUÍ
texto = ""
entero = 0
decimal = 0.0
activo = False

print(texto)
print(entero)
print(decimal)
print(activo)

# Prueba simple
assert isinstance(texto, str) and isinstance(entero, int)
````

### Ejercicio 4: Combinar texto
**Enunciado:** crea dos variables de texto y combínalas en una sola frase.

**Ejemplo esperado:**
```
Aprender Python es divertido.
```

**Pista:** puedes usar `+` o un f-string.


````python
# TU CÓDIGO AQUÍ
parte1 = ""
parte2 = ""

frase = f"{parte1} {parte2}".strip()
print(frase)

# Prueba simple
assert isinstance(frase, str) and len(frase) > 0
````

## ¿Qué es una variable?

Una variable es un nombre que guarda un valor para poder usarlo después.
Es como una caja con etiqueta: puedes guardar un dato y consultarlo cuando lo necesites.
En Python el tipo puede cambiar porque el lenguaje es de tipado dinámico.


## Ejemplo guiado


````python
nombre = "Daniela"
edad = 29
activo = True
print(nombre, edad, activo)
````

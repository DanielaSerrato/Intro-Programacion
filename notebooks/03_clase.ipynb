# Clase 3: Modelo conceptual (ER), reglas y diccionario de datos

## Objetivos
- Identificar **entidades**, **atributos** y **relaciones**.
- Definir **cardinalidades** y reglas de negocio.
- Elaborar un **diccionario de datos** básico.


## Modelo conceptual (ER)
El modelo Entidad–Relación (ER) describe *qué información existe* y *cómo se conecta*.

- **Entidad:** objeto del mundo real (Cliente, Producto).
- **Atributo:** característica de la entidad (nombre, precio).
- **Relación:** vínculo entre entidades (Cliente realiza Pedido).


## Ejemplo guiado: tienda en línea
**Entidades principales:** Cliente, Pedido, Producto.

**Relaciones:**
- Un **Cliente** realiza **muchos Pedidos**.
- Un **Pedido** incluye **muchos Productos** (vía DetallePedido).


### Reglas de negocio (guía)
1. Todo pedido debe pertenecer a un cliente.
2. Un producto puede aparecer en muchos pedidos.
3. Un pedido debe tener al menos un detalle.


### Diccionario de datos (extracto)
| Entidad | Atributo | Tipo | Descripción |
| --- | --- | --- | --- |
| Cliente | cliente_id | entero | Identificador único |
| Cliente | nombre | texto | Nombre completo |
| Pedido | pedido_id | entero | Identificador del pedido |
| Pedido | fecha | fecha | Día del pedido |
| Producto | producto_id | entero | Identificador del producto |
| Producto | precio | decimal | Precio unitario |


```python
diccionario = [
    {"entidad": "Cliente", "atributo": "cliente_id", "tipo": "entero", "descripcion": "Identificador único"},
    {"entidad": "Cliente", "atributo": "nombre", "tipo": "texto", "descripcion": "Nombre completo"},
    {"entidad": "Pedido", "atributo": "pedido_id", "tipo": "entero", "descripcion": "Identificador del pedido"},
    {"entidad": "Producto", "atributo": "precio", "tipo": "decimal", "descripcion": "Precio unitario"},
]

for item in diccionario:
    print(f"{item['entidad']}.{item['atributo']} -> {item['tipo']}")
```

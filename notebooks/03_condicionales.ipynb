# Condicionales

Este capítulo introduce las decisiones en los programas mediante condicionales.
Aprenderás a usar `if`, `elif` y `else` para controlar el flujo.
Practicarás comparaciones y condiciones compuestas.
Estas ideas serán clave para ejercicios y proyectos más complejos.


## ¿Qué es un condicional?

Un condicional permite que el programa tome decisiones.
Evalúa una condición y elige qué camino seguir (`if`, `elif`, `else`).
Es como decir: si pasa esto, haz aquello; si no, haz otra cosa.


## Ejemplo guiado


````python
edad = 18
if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
````

### Ejercicio 1: Número positivo
**Enunciado:** crea una variable `numero` y muestra si es positivo, negativo o cero.

**Ejemplo esperado:**
```
El número es positivo
```

**Pista:** usa `if`, `elif` y `else`.


````python
# TU CÓDIGO AQUÍ
numero = 0

if numero > 0:
    mensaje = "El número es positivo"
elif numero < 0:
    mensaje = "El número es negativo"
else:
    mensaje = "El número es cero"

print(mensaje)

# Prueba simple
assert mensaje in {"El número es positivo", "El número es negativo", "El número es cero"}
````

### Ejercicio 2: Par o impar
**Enunciado:** crea una variable `numero` y muestra si es par o impar.

**Ejemplo esperado:**
```
El número es par
```

**Pista:** usa el operador `%`.


````python
# TU CÓDIGO AQUÍ
numero = 0

if numero % 2 == 0:
    resultado = "El número es par"
else:
    resultado = "El número es impar"

print(resultado)

# Prueba simple
assert resultado in {"El número es par", "El número es impar"}
````

### Ejercicio 3: Comparar dos números
**Enunciado:** crea `a` y `b` y muestra cuál es mayor. Si son iguales, indícalo.

**Ejemplo esperado:**
```
El mayor es 10
```

**Pista:** usa una cadena para el mensaje final.


````python
# TU CÓDIGO AQUÍ
a = 0
b = 0

if a > b:
    mensaje = f"El mayor es {a}"
elif b > a:
    mensaje = f"El mayor es {b}"
else:
    mensaje = "Son iguales"

print(mensaje)

# Prueba simple
assert isinstance(mensaje, str) and mensaje != ""
````

### Ejercicio 4: Condiciones compuestas
**Enunciado:** crea variables `edad` y `tiene_invitacion`.
La persona puede entrar si tiene 18 o más **y** tiene invitación.
Muestra el mensaje correspondiente.

**Ejemplo esperado:**
```
Puede entrar
```

**Pista:** usa `and`.


````python
# TU CÓDIGO AQUÍ
edad = 0
tiene_invitacion = False

if edad >= 18 and tiene_invitacion:
    mensaje = "Puede entrar"
else:
    mensaje = "No puede entrar"

print(mensaje)

# Prueba simple
assert mensaje in {"Puede entrar", "No puede entrar"}
````

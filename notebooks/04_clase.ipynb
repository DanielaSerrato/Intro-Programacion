# Clase 4: Modelo lógico (tablas, PK/FK e integridad)

## Objetivos
- Transformar un modelo conceptual ER en **tablas relacionales**.
- Identificar **claves primarias (PK)** y **claves foráneas (FK)**.
- Reconocer reglas de **integridad**.


## Del modelo ER al modelo lógico
Reglas básicas de transformación:

1. Cada **entidad** se convierte en una tabla.
2. Cada **atributo** se convierte en una columna.
3. Las **relaciones** se representan con claves foráneas.


## Ejemplo guiado: academia
**Entidades:** Estudiante, Curso, Inscripción.

**Modelo lógico:**
- **Estudiante**(estudiante_id PK, nombre, correo)
- **Curso**(curso_id PK, nombre, creditos)
- **Inscripcion**(inscripcion_id PK, estudiante_id FK, curso_id FK, fecha)


### Integridad referencial
La FK `estudiante_id` en **Inscripcion** debe existir en **Estudiante**.
Lo mismo para `curso_id` en **Curso**.


```python
estudiantes = [
    {"estudiante_id": 1, "nombre": "Sofía"},
    {"estudiante_id": 2, "nombre": "Mateo"},
]

inscripciones = [
    {"inscripcion_id": 10, "estudiante_id": 1, "curso_id": 100},
    {"inscripcion_id": 11, "estudiante_id": 3, "curso_id": 100},
]

estudiante_ids = {e["estudiante_id"] for e in estudiantes}

for inscripcion in inscripciones:
    if inscripcion["estudiante_id"] not in estudiante_ids:
        print(f"Error FK: estudiante_id {inscripcion['estudiante_id']} no existe")
```

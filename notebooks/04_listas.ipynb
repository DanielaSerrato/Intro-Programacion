# Listas


En este capítulo trabajarás con listas como colección de datos.
Aprenderás a crear, acceder y modificar elementos.
Explorarás métodos comunes y recorridos con ciclos.
Las listas serán esenciales para ejemplos más grandes.


## Ejemplo guiado


````python
frutas = ["manzana", "pera", "banano"]
print(frutas[1])
frutas.append("mango")
print(frutas)
````

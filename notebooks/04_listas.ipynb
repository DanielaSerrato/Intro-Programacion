# Listas

Una lista es una colección ordenada y modificable de elementos.
Permite almacenar múltiples valores dentro de una sola variable.

En esta lección aprenderás:

- Cómo crear una lista
- Cómo acceder a elementos por índice
- Cómo modificar listas
- Métodos comunes como append, remove, sort
- Cómo recorrer listas con un ciclo for


En este capítulo trabajarás con listas como colección de datos.
Aprenderás a crear, acceder y modificar elementos.
Explorarás métodos comunes y recorridos con ciclos.
Las listas serán esenciales para ejemplos más grandes.


## Ejemplo guiado


````python
frutas = ["manzana", "pera", "banano"]
print(frutas[1])
frutas.append("mango")
print(frutas)
````

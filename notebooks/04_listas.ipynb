## ¿Qué es una lista?

Una lista es una colección ordenada de elementos.
Puedes guardar varios valores en una sola variable y acceder por índice.
Las listas son mutables: puedes agregar, quitar o cambiar elementos.


## Ejemplo guiado


````python
frutas = ["manzana", "pera", "banano"]
print(frutas[1])
frutas.append("mango")
print(frutas)
````

## Ejercicios

Sigue el formato en cada ejercicio: completa el bloque **TU CÓDIGO AQUÍ** y revisa el ejemplo.


### Ejercicio 1: Crear una lista
**Enunciado:** crea una lista `animales` con tres animales y muéstrala.

**Ejemplo esperado:**
```
['perro', 'gato', 'pez']
```

**Pista:** usa corchetes `[]`.


````python
# TU CÓDIGO AQUÍ
animales = []

print(animales)

# Prueba simple
assert isinstance(animales, list) and len(animales) == 3
````

### Ejercicio 2: Acceder por índice
**Enunciado:** muestra el primer elemento de la lista `animales`.

**Ejemplo esperado:**
```
perro
```

**Pista:** el primer índice es `0`.


````python
# TU CÓDIGO AQUÍ
animales = ["perro", "gato", "pez"]

primer_animal = animales[0]
print(primer_animal)

# Prueba simple
assert primer_animal == animales[0]
````

### Ejercicio 3: Agregar elementos
**Enunciado:** agrega un nuevo animal a la lista `animales` usando `append`.

**Ejemplo esperado:**
```
['perro', 'gato', 'pez', 'loro']
```

**Pista:** usa `animales.append("loro")`.


````python
# TU CÓDIGO AQUÍ
animales = ["perro", "gato", "pez"]

animales.append("loro")
print(animales)

# Prueba simple
assert "loro" in animales
````

### Ejercicio 4: Eliminar elementos
**Enunciado:** elimina un elemento de la lista `animales` usando `remove`.

**Ejemplo esperado:**
```
['perro', 'pez']
```

**Pista:** `remove` elimina por valor.


````python
# TU CÓDIGO AQUÍ
animales = ["perro", "gato", "pez"]

animales.remove("gato")
print(animales)

# Prueba simple
assert "gato" not in animales
````

### Ejercicio 5: Tamaño de la lista
**Enunciado:** crea una lista `numeros` con 4 números y guarda su tamaño en `cantidad`.

**Ejemplo esperado:**
```
4
```

**Pista:** usa `len(numeros)`.


````python
# TU CÓDIGO AQUÍ
numeros = [1, 2, 3, 4]

cantidad = len(numeros)
print(cantidad)

# Prueba simple
assert cantidad == 4
````

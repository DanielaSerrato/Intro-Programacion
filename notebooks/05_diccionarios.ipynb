# Diccionarios

Este capítulo introduce los diccionarios como pares clave–valor.
Aprenderás a crear, acceder y actualizar información rápidamente.
Los diccionarios son esenciales para representar datos estructurados.


## ¿Qué es un diccionario?

Un diccionario guarda información en pares clave–valor.
La clave es como una etiqueta y el valor es el dato asociado.
Sirve para acceder rápido a información organizada.


## Ejemplo guiado
Nota: `.get()` permite evitar errores cuando una clave no existe.


````python
persona = {"nombre": "Daniela", "edad": 29}
print(persona["nombre"])
persona["edad"] = 30
print(persona)
````

## Ejercicios

Sigue el formato en cada ejercicio: completa el bloque **TU CÓDIGO AQUÍ** y revisa el ejemplo.


### Ejercicio 1: Crear un diccionario
**Enunciado:** crea un diccionario `persona` con las claves `nombre` y `edad`.

**Ejemplo esperado:**
```
{'nombre': 'Ana', 'edad': 20}
```

**Pista:** usa llaves `{}`.


````python
# TU CÓDIGO AQUÍ
persona = {}

print(persona)

# Prueba simple
assert "nombre" in persona and "edad" in persona
````

### Ejercicio 2: Acceder a un valor
**Enunciado:** muestra solo el valor de la clave `edad` del diccionario `persona`.

**Ejemplo esperado:**
```
20
```

**Pista:** usa `persona["edad"]`.


````python
# TU CÓDIGO AQUÍ
persona = {"nombre": "Ana", "edad": 20}

edad = persona["edad"]
print(edad)

# Prueba simple
assert edad == 20
````

### Ejercicio 3: Agregar una clave
**Enunciado:** agrega la clave `ciudad` al diccionario `persona`.

**Ejemplo esperado:**
```
{'nombre': 'Ana', 'edad': 20, 'ciudad': 'Cali'}
```

**Pista:** usa `persona["ciudad"] = "Cali"`.


````python
# TU CÓDIGO AQUÍ
persona = {"nombre": "Ana", "edad": 20}

persona["ciudad"] = ""
print(persona)

# Prueba simple
assert "ciudad" in persona
````

### Ejercicio 4: Usar `.get()`
**Enunciado:** usa `.get()` para leer la clave `telefono` en el diccionario `persona`.
Si no existe, muestra el valor por defecto "Sin teléfono".

**Ejemplo esperado:**
```
Sin teléfono
```

**Pista:** `persona.get("telefono", "Sin teléfono")`.


````python
# TU CÓDIGO AQUÍ
persona = {"nombre": "Ana", "edad": 20}

telefono = persona.get("telefono", "Sin teléfono")
print(telefono)

# Prueba simple
assert telefono == "Sin teléfono"
````

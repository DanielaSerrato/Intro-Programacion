# Diccionarios


Este capítulo introduce los diccionarios como pares clave–valor.
Aprenderás a crear, acceder y actualizar información rápidamente.
Verás cómo agregar, eliminar y recorrer elementos.
Los diccionarios son esenciales para representar datos estructurados.


## Ejemplo guiado


````python
persona = {"nombre": "Daniela", "edad": 29}
print(persona["nombre"])
persona["edad"] = 30
print(persona)
````

Nota: `.get()` permite evitar errores cuando una clave no existe.

Diccionarios

Un diccionario es una colección de pares clave–valor.
Cada clave está asociada a un valor específico y permite acceder a la
información rápidamente.

En esta lección aprenderás:

- Cómo crear diccionarios
- Cómo acceder y actualizar valores
- Cómo agregar y eliminar elementos
- Cómo recorrer claves y valores


## Ejemplo

Creamos un diccionario que representa a una persona.
Accedemos a valores y actualizamos información.


````python
persona = {
    "nombre": "Daniela",
    "edad": 29,
    "ciudad": "Bogotá"
}

print(persona["nombre"])

persona["edad"] = 30
print(persona)
````

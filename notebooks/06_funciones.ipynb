## ¿Qué es una función?

Una función es un bloque de código con un nombre.
Recibe datos de entrada (parámetros) y puede devolver un resultado.


## Ejemplo guiado


````python
def saludar(nombre):
    return f"Hola, {nombre}!"

mensaje = saludar("Laura")
print(mensaje)
````

Sigue el formato en cada ejercicio: completa el bloque **TU CÓDIGO AQUÍ** y revisa el ejemplo.


### Ejercicio 1: Función de suma
**Enunciado:** crea una función `sumar(a, b)` que devuelva la suma de dos números.

**Ejemplo esperado:**
```
sumar(3, 4) -> 7
```

**Pista:** usa `return a + b`.


````python
# TU CÓDIGO AQUÍ

def sumar(a, b):
    return a + b

# Pruebas simples
assert sumar(3, 4) == 7
assert sumar(-1, 1) == 0
````

### Ejercicio 2: Función de mensaje
**Enunciado:** crea una función `crear_mensaje(nombre, ciudad)` que devuelva una frase con ambos datos.

**Ejemplo esperado:**
```
"Ana vive en Lima."
```

**Pista:** usa f-strings.


````python
# TU CÓDIGO AQUÍ

def crear_mensaje(nombre, ciudad):
    return f"{nombre} vive en {ciudad}."

# Prueba simple
assert crear_mensaje("Ana", "Lima") == "Ana vive en Lima."
````

### Ejercicio 3: Función con booleanos
**Enunciado:** crea una función `es_mayor_de_edad(edad)` que devuelva `True` si `edad` es 18 o más.

**Ejemplo esperado:**
```
es_mayor_de_edad(20) -> True
```

**Pista:** la función puede devolver una comparación directamente.


````python
# TU CÓDIGO AQUÍ

def es_mayor_de_edad(edad):
    return edad >= 18

# Pruebas simples
assert es_mayor_de_edad(18) is True
assert es_mayor_de_edad(10) is False
````

### Ejercicio 4: Cálculo de área
**Enunciado:** crea una función `area_rectangulo(base, altura)` que devuelva el área.

**Ejemplo esperado:**
```
area_rectangulo(5, 2) -> 10
```

**Pista:** el área es `base * altura`.


````python
# TU CÓDIGO AQUÍ

def area_rectangulo(base, altura):
    return base * altura

# Pruebas simples
assert area_rectangulo(5, 2) == 10
assert area_rectangulo(0, 3) == 0
````

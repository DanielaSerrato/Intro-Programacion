# Loops (Ciclos)

Un loop permite repetir una acción múltiples veces.  
Python incluye dos estructuras principales:

- `for`: ideal para recorrer listas, cadenas y colecciones
- `while`: se ejecuta mientras una condición sea verdadera

En esta lección aprenderás a usar ambos ciclos con ejemplos claros.


En este capítulo practicarás ciclos for y while para repetir acciones.
Aprenderás a recorrer listas y controlar iteraciones.
Verás cómo automatizar tareas repetitivas con pocos comandos.
Este tema es clave para trabajar con colecciones de datos.


## Ejemplo guiado


````python
frutas = ["manzana", "pera", "mango"]
for fruta in frutas:
    print(fruta)
````

Nota: Usa `break` para detener un loop y `continue` para saltar a la siguiente iteración.

## ¿Qué es un loop (ciclo)?

Un loop repite un bloque de código varias veces.
- `for`: repite sobre una colección o un rango.
- `while`: repite mientras una condición sea verdadera.


## Ejemplo guiado


````python
# Loop con for
numeros = [1, 2, 3]
for numero in numeros:
    print(numero)

# Loop con while
contador = 0
while contador < 3:
    print("Contador:", contador)
    contador += 1
````

## Ejercicios

Sigue el formato en cada ejercicio: completa el bloque **TU CÓDIGO AQUÍ** y revisa el ejemplo.


### Ejercicio 1: Recorrer una lista
**Enunciado:** imprime cada elemento de la lista `frutas`.

**Ejemplo esperado:**
```
manzana
pera
banano
```

**Pista:** usa `for`.


````python
# TU CÓDIGO AQUÍ
frutas = ["manzana", "pera", "banano"]

for fruta in frutas:
    print(fruta)

# Prueba simple
assert len(frutas) == 3
````

### Ejercicio 2: Sumatoria con for
**Enunciado:** calcula la suma de la lista `numeros` usando un acumulador.

**Ejemplo esperado:**
```
6
```

**Pista:** inicia `suma = 0` y luego suma cada número.


````python
# TU CÓDIGO AQUÍ
numeros = [1, 2, 3]

suma = 0
for numero in numeros:
    suma += numero

print(suma)

# Prueba simple
assert suma == 6
````

### Ejercicio 3: Contar con while
**Enunciado:** usa un `while` para imprimir los números del 1 al 3.

**Ejemplo esperado:**
```
1
2
3
```

**Pista:** incrementa el contador dentro del ciclo.


````python
# TU CÓDIGO AQUÍ
contador = 1

while contador <= 3:
    print(contador)
    contador += 1

# Prueba simple
assert contador == 4
````

### Ejercicio 4: Rango con for
**Enunciado:** imprime los números del 0 al 4 usando `range`.

**Ejemplo esperado:**
```
0
1
2
3
4
```

**Pista:** `range(5)` genera 0, 1, 2, 3, 4.


````python
# TU CÓDIGO AQUÍ
for numero in range(5):
    print(numero)

# Prueba simple
assert list(range(5)) == [0, 1, 2, 3, 4]
````

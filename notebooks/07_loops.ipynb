# Loops (Ciclos)


En este capítulo practicarás ciclos for y while para repetir acciones.
Aprenderás a recorrer listas y controlar iteraciones.
Verás cómo automatizar tareas repetitivas con pocos comandos.
Este tema es clave para trabajar con colecciones de datos.


## Ejemplo guiado
Nota: Usa `break` para detener un loop y `continue` para saltar a la siguiente iteración.


````python
frutas = [\"manzana\", \"pera\", \"mango\"]\nfor fruta in frutas:\n    print(fruta)\n
````

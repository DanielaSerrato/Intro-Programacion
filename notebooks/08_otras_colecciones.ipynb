## ¿Qué son las tuplas y los sets?

Las tuplas son listas que no se pueden modificar (inmutables).
Los sets son colecciones sin elementos duplicados.
Se usan cuando quieres datos fijos (tuplas) o eliminar repeticiones (sets).

# Otras colecciones en Python: Tuplas y Conjuntos (sets)

En este capítulo conocerás dos colecciones adicionales: **tuplas** y **sets**.
Verás cuándo conviene usar cada una y cómo crear ejemplos sencillos.


## ¿Qué son las tuplas y los sets?

- **Tupla:** colección ordenada e inmutable.
- **Set (conjunto):** colección no ordenada y sin elementos repetidos.


## Ejemplo guiado


````python
# Tupla
coordenadas = (10, 20)
print(coordenadas[0])

# Set
colores = {"rojo", "azul", "rojo"}
print(colores)
````

### Ejercicio 1: Crear una tupla
**Enunciado:** crea una tupla `datos` con tu nombre, edad y ciudad.

**Ejemplo esperado:**
```
('Ana', 20, 'Cali')
```

**Pista:** usa paréntesis `()`.


````python
# TU CÓDIGO AQUÍ
datos = ()

print(datos)

# Prueba simple
assert isinstance(datos, tuple) and len(datos) == 3
````

### Ejercicio 2: Acceder a una tupla
**Enunciado:** muestra el segundo elemento de la tupla `datos`.

**Ejemplo esperado:**
```
20
```

**Pista:** el índice 1 corresponde al segundo elemento.


````python
# TU CÓDIGO AQUÍ
datos = ("Ana", 20, "Cali")

segundo = datos[1]
print(segundo)

# Prueba simple
assert segundo == 20
````

### Ejercicio 3: Eliminar duplicados con set
**Enunciado:** crea un set `numeros` a partir de una lista con números repetidos y muestra el resultado.

**Ejemplo esperado:**
```
{1, 2, 3}
```

**Pista:** puedes usar `set(lista)`.


````python
# TU CÓDIGO AQUÍ
lista = [1, 2, 2, 3, 3]

numeros = set(lista)
print(numeros)

# Prueba simple
assert len(numeros) == 3
````

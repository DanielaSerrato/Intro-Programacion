````python
estudiantes = [
    {"nombre": "Ana", "edad": 20},
    {"nombre": "Luis", "edad": 25},
    {"nombre": "Sofia", "edad": 22},
]

def mayores_de_edad(lista):
    return [e for e in lista if e["edad"] >= 21]

mayores_de_edad(estudiantes)
````

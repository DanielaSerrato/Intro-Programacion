## Ejemplo guiado


````python
# Lista de estudiantes y función de resumen
estudiantes = [
    {"nombre": "Ana", "edad": 20},
    {"nombre": "Luis", "edad": 22},
]

def promedio_edades(lista):
    total = 0
    for estudiante in lista:
        total += estudiante["edad"]
    return total / len(lista)

print(promedio_edades(estudiantes))
````

### Ejercicio 1: Filtrar mayores de edad
**Enunciado:** usando la lista `personas`, crea una nueva lista con los nombres de quienes tienen 18 o más.

**Ejemplo esperado:**
```
['Ana', 'Luis']
```

**Pista:** usa un `for` y una condición.


````python
# TU CÓDIGO AQUÍ
personas = [
    {"nombre": "Ana", "edad": 20},
    {"nombre": "Paz", "edad": 16},
    {"nombre": "Luis", "edad": 22},
]

mayores = []
for persona in personas:
    if persona["edad"] >= 18:
        mayores.append(persona["nombre"])

print(mayores)

# Prueba simple
assert mayores == ["Ana", "Luis"]
````

### Ejercicio 2: Función de promedio
**Enunciado:** crea una función `promedio(numeros)` que devuelva el promedio de una lista.

**Ejemplo esperado:**
```
promedio([2, 4, 6]) -> 4
```

**Pista:** suma los números y divide por `len(numeros)`.


````python
# TU CÓDIGO AQUÍ

def promedio(numeros):
    total = 0
    for numero in numeros:
        total += numero
    return total / len(numeros)

# Pruebas simples
assert promedio([2, 4, 6]) == 4
assert promedio([1, 1]) == 1
````

### Ejercicio 3: Contador de palabras
**Enunciado:** crea un diccionario `contador` con el número de veces que aparece cada palabra en `palabras`.

**Ejemplo esperado:**
```
{'hola': 2, 'python': 1}
```

**Pista:** si una palabra no existe en el diccionario, empieza en 0.


````python
# TU CÓDIGO AQUÍ
palabras = ["hola", "python", "hola"]

contador = {}
for palabra in palabras:
    contador[palabra] = contador.get(palabra, 0) + 1

print(contador)

# Prueba simple
assert contador == {"hola": 2, "python": 1}
````

### Ejercicio 4: Mini-menú sin input
**Enunciado:** crea una función `accion(opcion)` que devuelva un mensaje según el número:
- 1: "Agregar"
- 2: "Mostrar"
- 3: "Salir"
- otro: "Opción inválida"

**Ejemplo esperado:**
```
accion(2) -> "Mostrar"
```

**Pista:** usa condicionales.


````python
# TU CÓDIGO AQUÍ

def accion(opcion):
    if opcion == 1:
        return "Agregar"
    if opcion == 2:
        return "Mostrar"
    if opcion == 3:
        return "Salir"
    return "Opción inválida"

# Pruebas simples
assert accion(2) == "Mostrar"
assert accion(9) == "Opción inválida"
````

# Proyecto Final: Sistema de Gestión de Usuarios

En este proyecto crearás una mini-aplicación que permita:
- Registrar usuarios
- Mostrar usuarios
- Buscar por nombre

Este proyecto integra todo lo aprendido en el curso.


## Datos iniciales

Usaremos una lista de diccionarios para guardar usuarios.


```python
# Lista principal de usuarios
usuarios = []
```

## Paso 1: Registrar usuarios

Crea una función `registrar_usuario` que reciba `nombre`, `edad` y `ciudad`,
cree un diccionario y lo agregue a la lista `usuarios`.


```python
# TU CÓDIGO AQUÍ

def registrar_usuario(nombre, edad, ciudad):
    usuario = {"nombre": nombre, "edad": edad, "ciudad": ciudad}
    usuarios.append(usuario)
    return usuario

# Pruebas simples
usuarios.clear()
registrar_usuario("Ana", 20, "Lima")
registrar_usuario("Luis", 22, "Quito")

assert len(usuarios) == 2
assert usuarios[0]["nombre"] == "Ana"
```

## Paso 2: Mostrar usuarios

Crea una función `listar_usuarios` que devuelva una lista con los nombres de los usuarios.


```python
# TU CÓDIGO AQUÍ

def listar_usuarios():
    nombres = []
    for usuario in usuarios:
        nombres.append(usuario["nombre"])
    return nombres

# Pruebas simples
assert listar_usuarios() == ["Ana", "Luis"]
```

## Paso 3: Buscar usuario

Crea una función `buscar_usuario(nombre)` que devuelva el diccionario del usuario
si existe o `None` si no está.


```python
# TU CÓDIGO AQUÍ

def buscar_usuario(nombre):
    for usuario in usuarios:
        if usuario["nombre"].lower() == nombre.lower():
            return usuario
    return None

# Pruebas simples
assert buscar_usuario("Ana")["ciudad"] == "Lima"
assert buscar_usuario("NoExiste") is None
```

## Paso 4: Eliminar usuario

Crea una función `eliminar_usuario(nombre)` que elimine al usuario por nombre.
Debe devolver `True` si eliminó algo o `False` si no encontró el usuario.


```python
# TU CÓDIGO AQUÍ

def eliminar_usuario(nombre):
    indice = 0
    while indice < len(usuarios):
        if usuarios[indice]["nombre"].lower() == nombre.lower():
            usuarios.pop(indice)
            return True
        indice += 1
    return False

# Pruebas simples
assert eliminar_usuario("Luis") is True
assert eliminar_usuario("Luis") is False
```
